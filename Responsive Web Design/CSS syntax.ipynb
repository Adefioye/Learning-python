CSS (Cascading style sheets) is used for displaying texts or content made by HTML. CSS can be used to control color, fonts, size, positioning, transitions, spacing and decorations.

There are like 3 ways of applying CSS rules to a webpage:
- Using inline styling approach of using CSS rules in the "style" attribute of HTML elements
- Applying the CSS rules within the "style" tag of within the HTML document
- USing external style sheets to make CSS rules which are then referenced by the HTML document

CSS works by using a __selector__ to target an HTML element in the DOM(Document Object Model and then applying a collage of properties in order to determine how it is displayed.
